```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results = {}

# vgg, blurred architecture without processed images, blurred architecture with processed images,
# depth architecture without processed images, depth architecture with processed images, 
# edge architecture without processed images, edge architecture with processed images

results['test'] = [0.928, 0.741, 0.836, 0.903, 0.939, 0.873, 0.790]

results['augmented'] = [0.935, 0.641, 0.934, 0.871, 0.744, 0.83, 0.714]

results['adversarial'] = [0.52, 0.509, 0.673, 0.513, 0.586, 0.496, 0.443]
```

```python
label = ["VGG16", "Approach 1 without processed images", "Approach 1 with processed images",
        "Approach 2 without processed images", "Approach 2 with processed images", 
        "Approach 3 without processed images", "Approach 3 with processed images"]
```

```python
cmap = matplotlib.cm.get_cmap('Pastel1')
cmap
```

```python
colors = ["#D6DDCD", "#EED4C6", "#C7D1DD", "#EFE3B1", "#CCCCCC", "#DBCCE2", "#EEE2CE"]
fig, ax = plt.subplots(1,1,figsize=(15,10))
plt.barh(label, results['test'], height=0.75, color=colors, linewidth=2, edgecolor='dimgray')
plt.xticks(np.arange(0,1.01,0.1), fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(f"\nAccuracy", fontsize=25)
# plt.title("Test Set Accuracy\n", fontsize=28)
plt.tight_layout()
plt.savefig("../Documentation/Plots/test.png",format='png', dpi=1200)
plt.show()
```

```python
colors = ["#D6DDCD", "#EED4C6", "#C7D1DD", "#EFE3B1", "#CCCCCC", "#DBCCE2", "#EEE2CE"]
fig, ax = plt.subplots(1,1,figsize=(15,10))
plt.barh(label, results['adversarial'], height=0.75, color=colors, linewidth=2, edgecolor='dimgray')
plt.xticks(np.arange(0,1.01,0.1), fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(f"\nAccuracy", fontsize=25)
# plt.title("Adversarial Test Set Accuracy\n", fontsize=28)
plt.tight_layout()
plt.savefig("../Documentation/Plots/adversarial.png",format='png', dpi=1200)
plt.show()
```

```python
colors = ["#D6DDCD", "#EED4C6", "#C7D1DD", "#EFE3B1", "#CCCCCC", "#DBCCE2", "#EEE2CE"]
fig, ax = plt.subplots(1,1,figsize=(15,10))
plt.barh(label, results['augmented'], height=0.75, color=colors, linewidth=2, edgecolor='dimgray')
plt.xticks(np.arange(0,1.01,0.1), fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(f"\nAccuracy", fontsize=25)
# plt.title("Augmented Test Set Accuracy\n", fontsize=28)
plt.tight_layout()
plt.savefig("../Documentation/Plots/augmented.png",format='png', dpi=1200)
plt.show()
```
